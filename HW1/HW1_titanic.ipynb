# Разбиение по полу

```python
data = pd.read_csv('titanic.csv')
```

```python
male_data = data[data['Sex'] == 'male']
female_data = data[data['Sex'] == 'female']

#Удаление пустых строк
male_data = male_data.dropna().reset_index(drop=True)
female_data = female_data.dropna().reset_index(drop=True)
```

```python
male_data
```

```python
female_data
```

# Построение гистограммы

```python
# Распределение возраста мужчин
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(15, 5))

ax1.hist(male_data['Age'], bins=20, color='blue', alpha=0.5)
ax1.set_title('Распределение возраста мужчин')
ax1.set_xlabel('Возраст')
ax1.set_ylabel('Количество')

# Распределение возраста женщин
# plt.figure(figsize=(6, 5))
ax2.hist(female_data['Age'], bins=20, color='red', alpha=0.7)
ax2.set_title('Распределение возраста женщин')
ax2.set_xlabel('Возраст')
ax2.set_ylabel('Количество')

# Распределение возраста пассажиров
# plt.figure(figsize=(6, 5))
ax3.hist(data['Age'], bins=20, color='green', alpha=0.9)
ax3.set_title('Распределение возраста пассажиров')
ax3.set_xlabel('Возраст')
ax3.set_ylabel('Количество')
```

# Вычислите мин и макс, среднее, моду, медиану возраста пассажира каждого пола

```python
# min age male
min_male_age = male_data['Age'].min()
# max age male
max_male_age = male_data['Age'].max()
# moda age male
mode_male_age = male_data['Age'].mode().iloc[0]
# median age male
median_male_age = male_data['Age'].median()
# mean age male
mean_male_age = male_data['Age'].mean()
print(f"Минимальный возраст мужчин: {min_male_age},\nМаксимальный возраст мужчин: {max_male_age},\nМода возраста мужчин: {mode_male_age},\nМедиана возраста мужчин: {median_male_age},\nСредний возраст мужчин: {mean_male_age:.2f}")
```

```python
# min age male
min_female_age = female_data['Age'].min()
# max age male
max_female_age = female_data['Age'].max()
# moda age male
mode_female_age = female_data['Age'].mode().iloc[0]
# median age male
median_female_age = female_data['Age'].median()
# mean age male
mean_female_age = female_data['Age'].mean()
print(f"Минимальный возраст женщин: {min_male_age},\nМаксимальный возраст женщин: {max_male_age},\nМода возраста женщин: {mode_male_age},\nМедиана возраста женщин: {median_male_age},\nСредний возраст женщин: {mean_male_age:.2f}")
```

# 4. Визуализация возраста

```python
x_male = male_data['Age']
plt.hist([male_data['Age'], female_data['Age']], bins=10, color=['blue', 'red'], alpha=0.5)
plt.legend(['Мужчины', 'Женщины'])
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.title('Распределение возраста пассажиров')
plt.show()
```

```python
fig, axes = plt.subplots(2, 4, figsize=(25, 15))

male_data['Age'].plot(kind='line', ax=axes[0,0], color='blue', title='Линейный график')
axes[0,0].set_xlabel('Индекс')
axes[0,0].set_ylabel('Возраст')

male_data['Age'].plot(kind='bar', ax=axes[0,1], color='red', title='Столбчатая диаграмма', rot=90, fontsize=8)
axes[0,1].set_xlabel('Индекс')
axes[0,1].set_ylabel('Возраст')

male_data['Age'].plot(kind='hist', ax=axes[0,2], color='green', edgecolor='black', title='Гистограмма')
axes[0,2].set_ylabel('Количество')
axes[0,2].set_xlabel('Возраст')

male_data['Age'].plot(kind='density', ax=axes[0,3], color='purple', alpha=0.5, title='Плотность распределения')
axes[0, 3].set_xlabel('Возраст')
axes[0, 3].set_ylabel('Плотность')

male_data['Age'].plot(kind='area', ax=axes[1,0], color='purple', alpha=0.5, title='Диаграмма с областью')
axes[1,0].set_xlabel('Индекс')
axes[1,0].set_ylabel('Возраст')

male_data['Age'].sort_values().plot(kind='pie', ax=axes[1,1], radius=1.5, fontsize=8, title='Круговая диаграмма')

male_data.reset_index().plot.scatter(x='index', y='Age', s = 100, ax=axes[1,2], title='Диаграмма рассеяния')
axes[1,2].set_xlabel('Индекс')
axes[1,2].set_ylabel('Возраст')

male_data.reset_index().plot.hexbin(x='index', y='Age', ax=axes[1,3], gridsize=10, 
                    cmap='YlOrRd', title='Гексагональная тепловая карта')
axes[1,3].set_xlabel('Индекс')
axes[1,3].set_ylabel('Возраст')

fig.suptitle("Распределение возраста мужчин", fontsize=20, fontweight='bold')
plt.tight_layout()
plt.show()
```
